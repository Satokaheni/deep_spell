```python
def preprocess(x):
    # lowercase 
    x = x.lower()
    
    # replace contractions
    x = ' '.join([contractions[w] if w in contractions.keys() else w for w in x.split()])
    
    # remove punctuation
    x = x.translate(str.maketrans('', '', string.punctuation))
    
    return x
```

```python
preprocessed_books = [preprocess(x) for x in lines]
```

```python
with open('../data/bookcorpus/preprocess_p1.pkl', 'wb') as f:
    pickle.dump(preprocessed_books, f)
```
